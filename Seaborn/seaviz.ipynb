```python
import seaborn as sns
import matplotlib.pyplot as plt
```

SCATTER PLOT

We use this plot to visualize the relationship of two numerical data.

With help of this plot we can interpret if there is a correlation between the two points.


```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Plotting Simple Scatter PLot

sns.scatterplot(x='total_bill' , y='tip' , data=tips)
plt.title('Total_bill vs tip')

plt.show()
```

```python
# STYLING THE SCATTER PLOT
sns.scatterplot(x='total_bill' , y='tip' , data=tips , hue='day' , style='time' , palette='coolwarm' , size='size')
plt.title('Total_bill vs tip')

plt.show()
```

LINE PLOT

A line plot displays data points connected by straight line segments, typically used to visualize data over time or any other continuous variable.


```python
# Simple Line Plot
data = sns.load_dataset('fmri')
data.head()
```

```python
sns.lineplot(x='timepoint' , y='signal' , data=data)
plt.title('Timepoint vs signal')

plt.show()
```

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x='year' , y='passengers' , hue='month' , data=flight , markers=True , style='month' , dashes=False)
plt.show()
```

BARPLOT

We use barplot to represent the categorical  data with the rectangular bar.

Each bar lenght coresponds to the value it represent.

categories or Groups are represented on X axis.

Values are reprensented on y-axis.

```python
data = sns.load_dataset('tips')
data.head()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='day' , y='total_bill' ,hue='sex', data=data , palette='hls' , errorbar=None)
plt.title('day vs total_bill')
plt.show()

# hue is used to decide which column in the dataframe should be used for color encoding.
```

HISTOGRAM

Histogram is used to display the distribution of numerical data.

It consists of bar that shows the frequency of data within specific range or bins.

It is used to analyze patterns and trends in data. Identifying the distribution shape(positive skewed , right skewed , left skewed)

```python
tips = sns.load_dataset('tips')

sns.histplot(tips['total_bill'], bins=10, kde=True) # bin is choice of spacing and unit on x axis.
plt.title('Histogram of Total Bill')
plt.show()
```

```python
data=sns.load_dataset('penguins')
data.head()
```

```python
sns.histplot(data['flipper_length_mm'] , bins = 10 , kde=True)
plt.title('Hist of flipper_length')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
ages = [5, 8, 12, 14, 15, 20, 21, 22, 25, 30, 35, 40]

# Histogram with fewer bins
plt.hist(ages, bins=3, edgecolor='black')
plt.title('Histogram with 3 Bins')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Histogram with more bins
plt.hist(ages, bins=6, edgecolor='black')
plt.title('Histogram with 6 Bins')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

BOX PLOT

It is also known as box-whisker plot

It provide us the summary of datasets thorugh five statistics:

-> Mininum

-> First Quartile (Q1)

-> Median(Q2)

-> Third Quartile (Q3)

-> Maximun

It is useful to visualize skweness and potential outliers in the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')

plt.show()
```

HEATMAP

visualizes data in a matrix format, where values are represented by colors. It's often used to explore correlations or patterns in large datasets.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
flights = sns.load_dataset('flights')

# Pivoting the  table for heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers' , observed=False)
print(flights_pivot)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
flights = sns.load_dataset('flights')

# Pivot table for heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers' , observed=False)

# Heatmap
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Heatmap of Flight Passengers')
plt.show()
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.heatmap(data.corr(numeric_only=True) , annot=True , linewidths=5 , cmap='viridis')
plt.title('Correlation haetmap')
```
