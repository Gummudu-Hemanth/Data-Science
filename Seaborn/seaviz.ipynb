```python
import seaborn as sns
import matplotlib.pyplot as plt
```

SCATTER PLOT

We use this plot to visualize the relationship of two numerical data.

With help of this plot we can interpret if there is a correlation between the two points.


```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Plotting Simple Scatter PLot

sns.scatterplot(x='total_bill' , y='tip' , data=tips)
plt.title('Total_bill vs tip')

plt.show()
```

```python
# STYLING THE SCATTER PLOT
sns.scatterplot(x='total_bill' , y='tip' , data=tips , hue='day' , style='time' , palette='coolwarm' , size='size')
plt.title('Total_bill vs tip')

plt.show()
```

BARPLOT

We use barplot to represent the categorical  data with the rectangular bar.

Each bar lenght coresponds to the value it represent.

categories or Groups are represented on X axis.

Values are reprensented on y-axis.

```python
data = sns.load_dataset('tips')
data.head()
```

HISTOGRAM

Histogram is used to display the distribution of numerical data.

It consists of bar that shows the frequency of data within specific range or bins.

It is used to analyze patterns and trends in data. Identifying the distribution shape(positive skewed , right skewed , left skewed)

```python
tips = sns.load_dataset('tips')

sns.histplot(tips['total_bill'], bins=10, kde=True) # bin is choice of spacing and unit on x axis.
plt.title('Histogram of Total Bill')
plt.show()
```

```python
data=sns.load_dataset('penguins')
data.head()
```

```python
sns.histplot(data['flipper_length_mm'] , bins = 10 , kde=True)
plt.title('Hist of flipper_length')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
ages = [5, 8, 12, 14, 15, 20, 21, 22, 25, 30, 35, 40]

# Histogram with fewer bins
plt.hist(ages, bins=3, edgecolor='black')
plt.title('Histogram with 3 Bins')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Histogram with more bins
plt.hist(ages, bins=6, edgecolor='black')
plt.title('Histogram with 6 Bins')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

BOX PLOT

It is also known as box-whisker plot

It provide us the summary of datasets thorugh five statistics:

-> Mininum

-> First Quartile (Q1)

-> Median(Q2)

-> Third Quartile (Q3)

-> Maximun

It is useful to visualize skweness and potential outliers in the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')

plt.show()
```

HEATMAP

visualizes data in a matrix format, where values are represented by colors. It's often used to explore correlations or patterns in large datasets.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
flights = sns.load_dataset('flights')

# Pivoting the  table for heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers' , observed=False)
print(flights_pivot)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
flights = sns.load_dataset('flights')

# Pivot table for heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers' , observed=False)

# Heatmap
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Heatmap of Flight Passengers')
plt.show()
```

### SUMMARY OF PLOTS , WHEN TO USE AND CLASSIFYING IT INTO UNIVARIATE AND BIVARIATE 

UNIVARIATE:-> It is used to describe and summarize single variable.

BIVARIATE:-> It is used to explore the relationships between two variables.


### Scatter Plot
- **Type:** Numerical
- **Univariate/Bivariate:** Bivariate
- **Scenario:** Used to observe relationships and correlations between two continuous variables. It's helpful for identifying trends, clusters, and outliers.

### Line Plot
- **Type:** Numerical
- **Univariate/Bivariate:** Bivariate
- **Scenario:** Used to display trends over time or ordered categories. It is particularly useful for time series data and showing how data points change over a continuous interval.

### Bar Plot
- **Type:** Categorical
- **Univariate/Bivariate:** Univariate (can be extended to bivariate with hue)
- **Scenario:** Used to compare the count or frequency of categorical data. It can show the distribution of a single categorical variable or compare different categories.

### Histogram
- **Type:** Numerical
- **Univariate/Bivariate:** Univariate
- **Scenario:** Used to show the distribution of a single continuous variable. It's useful for understanding the frequency distribution and identifying patterns such as skewness or modality.

### Box Plot
- **Type:** Numerical
- **Univariate/Bivariate:** Univariate (can be extended to bivariate with categories)
- **Scenario:** Used to display the distribution of data based on a five-number summary (minimum, first quartile, median, third quartile, and maximum). It is helpful for detecting outliers and comparing distributions across categories.

### Heatmap
- **Type:** Numerical
- **Univariate/Bivariate:** Bivariate
- **Scenario:** Used to show the intensity of data at the intersection of two variables using color. It is useful for visualizing matrix-like data, such as correlation matrices or comparing values across two dimensions.

### Pair Plot
- **Type:** Numerical
- **Univariate/Bivariate:** Multivariate (combination of bivariate plots)
- **Scenario:** Used to visualize pairwise relationships in a dataset. It plots all possible scatter plots for a given set of variables, making it useful for exploratory data analysis to detect patterns and relationships.

### Count Plot
- **Type:** Categorical
- **Univariate/Bivariate:** Univariate (can be extended to bivariate with hue)
- **Scenario:** Used to show the count of observations in each categorical bin. It is similar to a bar plot but specifically for counting the frequency of categories in a dataset.

### Reg Plot
- **Type:** Numerical
- **Univariate/Bivariate:** Bivariate
- **Scenario:** Used to plot data points along with a regression line to show the relationship between two continuous variables. It helps in understanding the strength and direction of the relationship.
